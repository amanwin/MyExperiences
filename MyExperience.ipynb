```python
import pandas as pd
from datetime import datetime
from dateutil import relativedelta

#
dob = datetime(1988,10,21)
dt_of_starting_job = datetime(2012,9,26)
dt_today = datetime.now()
total_age = relativedelta.relativedelta(dt_today,dob)
total_exp = relativedelta.relativedelta(dt_today,dt_of_starting_job)

print("Today's date",datetime.now().date())
print('My age till today is: {} years {} months {} days'.format(total_age.years,total_age.months,total_age.days))
print('My total experience till today is: {} years {} months {} days'.format(total_exp.years,total_exp.months,total_exp.days))

#United Health Group
ust = datetime(2012,9,26)
uet = datetime(2015,12,30)
utime = relativedelta.relativedelta(uet,ust)

#Mahindra Comviva
mst = datetime(2016,1,4)
met = datetime(2017,3,31)
mtime = relativedelta.relativedelta(met,mst)

#Fidelity International
fst = datetime(2017,4,3)
fet = datetime(2019,10,16)
ftime = relativedelta.relativedelta(fet,fst)

#Macquarie Global Services Pvt. Ltd.
mqst = datetime(2019,10,17)
mqet = datetime(2022,4,25)
mqtime = relativedelta.relativedelta(mqet,mqst)

data = [['United Health Group','812167', ust, uet, 'Associate Software Engineer', 'Apps Devpr 1','{} years {} months {} days'.format(utime.years,utime.months,utime.days)],
        ['Mahindra Comviva','5345', mst, met, 'Senior Engineer', 'Senior Engineer','{} years {} months {} days'.format(mtime.years,mtime.months,mtime.days)],
       ['Fidelity International','A604854', fst, fet, 'Programmer Analyst', 'Programmer Analyst','{} years {} months {} days'.format(ftime.years,ftime.months,ftime.days)],
       ['Macquarie Global Services Pvt. Ltd.','94751', mqst, mqet, 'Assistant Manager', 'Assistant Manager','{} years {} months {} days'.format(mqtime.years,mqtime.months,mqtime.days)]]
df = pd.DataFrame(data, columns=['Company Name','Emp Id', 'Start Date', 'End Date', 'Position Joined', 'Last Position Held', 'Duration'])
df.head()
```
